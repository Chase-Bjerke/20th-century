## 02. Web Scrapping 
### Key Events of the 20th Century

```python
# Define the URL as a string
page_url = "https://en.wikipedia.org/wiki/Key_events_of_the_20th_century"
```

```python
# Fetch the page with requests
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
                  "AppleWebKit/537.36 (KHTML, like Gecko) "
                  "Chrome/120.0.0.0 Safari/537.36"
}
```

```python
response = requests.get(page_url, headers=headers)
response.raise_for_status()  # stop if the request fails
```

```python
# Parse with BeautifulSoup
soup = BeautifulSoup(response.text, "html.parser")
print(soup.title.get_text())
```

```python
# Get all text from the page
text = soup.get_text()
```

```python
# Save to file
with open("20th_century_events.txt", "w", encoding="utf-8") as f:
    f.write(text)
```
