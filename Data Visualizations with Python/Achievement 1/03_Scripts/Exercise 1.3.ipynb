## 03. Pets in 5 Years

```python
df = pd.DataFrame({
    'pet': ['Dog', 'Cat', 'Rabbit', 'Parrot', 'Hamster'],
    'age': np.random.randint(1, 15, size=5)   # random ages between 1 and 15
})

# Lambda function
df['age_plus_five'] = df['age'].apply(lambda x: x + 5)

df
```

## 04. Indoor Vs Outdoor Pets

```python
df = pd.DataFrame({
    'pet': ['Dog', 'Cat', 'Rabbit', 'Parrot', 'Hamster'],
    'age': np.random.randint(1, 15, size=5)   # random ages between 1 and 15
})

# Lambda function
df['age_plus_five'] = df['age'].apply(lambda x: x + 5)

# Indoor/Outdoor pet flags
df['location'] = df['pet'].apply(
    lambda x: 'Indoor' if x in ['Rabbit', 'Hamster', 'Parrot'] else 'Outdoor'
)
df
```
